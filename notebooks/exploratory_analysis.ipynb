```python
from pathlib import Path

import pandas as pd

processed_root = Path('..') / 'data' / 'processed'
rows = []
for split_dir in sorted(processed_root.iterdir()):
    if not split_dir.is_dir():
        continue
    for class_dir in sorted(split_dir.iterdir()):
        if not class_dir.is_dir():
            continue
        rows.append(
            {
                'split': split_dir.name,
                'class_name': class_dir.name,
                'image_count': len([path for path in class_dir.iterdir() if path.is_file()]),
            }
        )

summary = pd.DataFrame(rows)
summary.sort_values(['split', 'class_name'])
```

```python
import matplotlib.pyplot as plt
from PIL import Image

sample_root = Path('..') / 'data' / 'samples'
sample_paths = sorted([path for path in sample_root.iterdir() if path.is_file()])[:4]

if sample_paths:
    figure, axes = plt.subplots(1, len(sample_paths), figsize=(4 * len(sample_paths), 4))
    axes = [axes] if len(sample_paths) == 1 else axes
    for axis, sample_path in zip(axes, sample_paths):
        axis.imshow(Image.open(sample_path).convert('RGB'))
        axis.set_title(sample_path.name)
        axis.axis('off')
    plt.tight_layout()
else:
    print('Add files under data/samples to preview example inputs.')
```

```python
evaluation_root = Path('..') / 'data' / 'processed' / 'evaluation'
metric_files = sorted(evaluation_root.glob('*_metrics.json'))
report_files = sorted(evaluation_root.glob('*_classification_report.csv'))

print('Metrics files:')
for metric_file in metric_files:
    print(' -', metric_file.name)

print('\nClassification report files:')
for report_file in report_files:
    print(' -', report_file.name)
```
